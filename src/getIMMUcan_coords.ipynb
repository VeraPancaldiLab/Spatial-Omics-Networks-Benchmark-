```python
IMMUcan_path = '../../IMMUcan_data/data/IMC/'



def get_matching_files(directory, pattern):
    matching_files = []
    # Use os.walk to traverse the directory and its subdirectories
    for root, dirs, files in os.walk(directory):
        # Use glob to match the pattern in the current directory
        for file in glob.glob(os.path.join(root, pattern)):
            matching_files.append(file)
    return matching_files

# Example usage:
file_pattern = '*cell_metadata*.csv'  # Change this pattern as needed
txt_files = get_matching_files(IMMUcan_path, file_pattern)
```

```python
import pandas as pd
import csv

All_Cells = []

for file in txt_files[-5:]:
    df = pd.read_csv(file)
    metadata = df[['Pos_X','Pos_Y','sample_id','tumor_patches','ROI']]
    metadata['TLS'] = False
    metadata[metadata['ROI']==4] ['TLS'] = True
    All_Cells.append(metadata)
AllCells = pd.concat(All_Cells)
```
